```python
import pandas as pd
import json
import math
import re
```

```python
data = pd.read_csv("metabolite_name_map.txt", sep="\t")
data_alt = pd.read_csv("metabolite_name_map_alt.txt", sep="\t")
```

```python
d1 = {}
d2 = {}
for i, r in data_alt.iterrows():
    d1[r[0]] = {"name": r[0], "alt_name": r[1], "hmdb_id": r[3], "kegg_id": r[2], "smiles": r[4]}
    d2[r[1]] = {"name": r[0], "alt_name": r[1], "hmdb_id": r[3], "kegg_id": r[2], "smiles": r[4]}
```

```python
d = {}
for i, r in data.iterrows():
    name = r[0]
    hmdb = r[1]
    chebi = r[2]
    kegg = r[3]
    smiles = r[4]
    
    try: 
        if math.isnan(chebi):
            chebi = ""
        else:
            chebi = str(int(chebi))
    except:
        pass
    try: 
        if math.isnan(hmdb):
            if name in d1 and not Math.isnan(d1[name]['hmdb_id']):
                hmdb = d1[name]['hmdb_id']
            elif name in d2 and d2[name]['hmdb_id'] != None:
                hmdb = d2[name]['hmdb_id']
            if math.isnan(hmdb):
                hmdb = ""
    except:
        pass
    try: 
        if math.isnan(kegg):
            if name in d1 and d1[name]['kegg_id'] != None:
                kegg = d1[name]['kegg_id']
            elif name in d2 and d2[name]['kegg_id'] != None:
                kegg = d2[name]['kegg_id']
            if math.isnan(kegg):
                kegg = ""
    except:
        pass
    try: 
        if math.isnan(smiles):
            if name in d1 and d1[name]['smiles'] != None:
                smiles = d1[name]['smiles']
            elif name in d2 and d2[name]['smiles'] != None:
                smiles = d2[name]['smiles']
            if math.isnan(smiles):
                smiles = ""
    except:
        pass
    _alt_name = ""
    
    if name in d1:
        _name = d1[name]["name"]
        _alt_name = d1[name]["alt_name"]
    elif name in d2:
        _name = d2[name]["name"]
        _alt_name = d2[name]["alt_name"]
    else:
        _name = name
        _alt_name = ""

    _name = re.sub(r'\W+', '', _name)
    _alt_name = re.sub(r'\W+', '', _alt_name)
        
    if _alt_name != "":
        d[_name.lower()] = {"name": _alt_name, "hmdb_id": hmdb, "kegg_id": kegg, "chebi_id": chebi, "smiles": smiles}
        d[_alt_name.lower()] = {"name": _alt_name, "hmdb_id": hmdb, "kegg_id": kegg, "chebi_id": chebi, "smiles": smiles}
    else:
        d[_name.lower()] = {"name": _name, "hmdb_id": hmdb, "kegg_id": kegg, "chebi_id": chebi, "smiles": smiles}
```
